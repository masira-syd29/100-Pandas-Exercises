##Ex1 - Filtering and Sorting Data


Step 1. Import the necessary libraries

```python
import pandas as pd
```

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
```

Step 4. How many products cost more than $10.00?


```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex= False).astype(float)
```

```python
mask_10 = chipo['item_price'] > 10.00
mask_10
```

```python
count_products = chipo[mask_10].shape[0]
count_products
```

```python
print(f'There are {count_products} products that cost more than $10.00')
```

```python
chipo.loc[(chipo['choice_description'] == '[Diet Coke]') & (chipo['item_name'] == 'Canned Soda')]
```

```python
chipo['product_price'] = chipo['item_price'] / chipo['quantity']
chipo
```

```python
chipo.loc[(chipo['choice_description'] == '[Diet Coke]') & (chipo['item_name'] == 'Canned Soda')]
```

```python
filtered_chipo = chipo.drop_duplicates(['item_name', 'choice_description'])
filtered_chipo
```

```python
filtered_chipo = filtered_chipo.loc[filtered_chipo['product_price'] > 10.00, ['item_name', 'choice_description', 'product_price']].reset_index(drop= True)
filtered_chipo
```

```python
print(f'There are {filtered_chipo.shape[0]} products that cost more than $10.00')
```

Step 5. What is the price of each item?


```python
filtered_chipo[['item_name', 'choice_description', 'product_price']]
filtered_chipo
```

Step 6. Sort by the name of the item


```python
chipo.item_name.sort_values()
```

```python
chipo.sort_values(by='item_name')
```

Step 7. What was the quantity of the most expensive item ordered?


```python
expensive_item_quantity = chipo.loc[chipo['item_price'].idxmax()]['quantity']
expensive_item_quantity
```

```python
print(f'The quantity of the most expensive item ordered is {expensive_item_quantity}')
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl = chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl']['quantity'].sum()
veggie_salad_bowl
```

```python
print(f'Veggie Salad Bowls were ordered {veggie_salad_bowl} times.')
```

Step 9. How many times did someone order more than one Canned Soda?


```python
canned_soda_orders = chipo[(chipo['item_name']=='Canned Soda') & (chipo['quantity']>1)].shape[0]
canned_soda_orders
```

```python
print(f'{canned_soda_orders} times people ordered more than one Canned Soda.')
```
