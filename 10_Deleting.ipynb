##Iris

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```


Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data).

Step 3. Assign it to a variable called

```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
iris
```

Step 4. Create columns for the dataset

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris.head()
```

Step 5. Is there any missing value in the dataframe?

```python
pd.isnull(iris).sum()
```

Step 6. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris.iloc[10:30, 2:3] = np.nan
iris.head(20)
```

Step 7. Good, now lets substitute the NaN values to 1.0

```python
iris.petal_length.fillna(1, inplace=True)
iris
```

Step 8. Now let's delete the column class

```python
del iris['class']
iris.head()
```

Step 9. Set the first 3 rows as NaN

```python
iris.loc[0:3, :] = np.nan
iris.head()
```

Step 10. Delete the rows that have NaN

```python
iris = iris.dropna(how='any')
iris.head()
```

Step 11. Reset the index so it begins with 0 again

```python
iris = iris.reset_index(drop=True)
iris.head()
```

##Wine

**Introduction:**

This exercise is a adaptation from the UCI Wine dataset. The only pupose is to practice deleting data with pandas.

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data).

Step 3. Assign it to a variable called wine

```python
wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
wine.head()
```

Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine = wine.drop(wine.columns[[0,3,6,8,10,12,13]], axis=1)
wine.head()
```

Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):

1. alcohol
2. malic_acid
3. alcalinity_of_ash
4. magnesium
5. flavanoids
6. proanthocyanins
7. hue

```python
wine.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wine.head()
```

Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.iloc[0:3, 0] = np.nan
wine.head()
```

Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.iloc[2:4, 3] = np.nan
wine.head()
```

Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
# wine.alcohol.fillna(10, inplace=True)
# wine.magnesium.fillna(100, inplace=True)
# wine.head()
wine = wine.fillna({'alcohol' : 10, 'magnesium':100})

wine.head()
```

Step 9. Count the number of missing values

```python
wine.isnull().sum()
```

Step 10. Create an array of 10 random numbers up until 10

```python
random = np.random.randint(10, size=10)
random
```

Step 11. Use random numbers you generated as an index and assign NaN value to each of cell.

```python
wine.loc[random, 'alcohol'] = np.nan
wine.head(10)
# The Modern, Single-Step Solution
# Syntax: df.loc[row_indexer, column_name] = value
```

Step 12. How many missing values do we have?

```python
wine.isnull().sum()
```

Step 13. Delete the rows that contain missing values

```python
wine = wine.dropna(axis=0, how='any')
wine.head()
```

Step 14. Print only the non-null values in alcohol

```python
mask = wine.alcohol.notnull()
mask
```

```python
wine.alcohol[mask]
```

Step 15. Reset the index, so it starts with 0 again

```python
wine = wine.reset_index(drop=True)
wine.head()
```
