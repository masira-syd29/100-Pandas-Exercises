##Student Alcohol Consumption

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv).


Step 3. Assign it to a variable called df.

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv')
df.head()
```

Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column


```python
student_alcohol = df.loc[:, 'school': 'guardian']
student_alcohol.head()
```

Step 5. Create a lambda function that will capitalize strings.


```python
capitalizer = lambda x : x.capitalize()
capitalizer
```

Step 6. Capitalize both Mjob and Fjob


```python
student_alcohol['Mjob'].apply(capitalizer)
student_alcohol['Fjob'].apply(capitalizer)
```

Step 7. Print the last elements of the data set.


```python
student_alcohol.tail()
```

Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.


```python
student_alcohol['Mjob'] = student_alcohol['Mjob'].apply(capitalizer)
student_alcohol['Fjob'] = student_alcohol['Fjob'].apply(capitalizer)
student_alcohol.tail()
```

Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)


```python
def majority(x):
  if x > 17 : return True
  else: return False
```

```python
student_alcohol['legal_drinker'] = student_alcohol['age'].apply(majority)
student_alcohol.head()
```

Step 10. Multiply every number of the dataset by 10.


```python
def times10(x):
  if type(x) is int:
    return 10 * x
  return x
```

```python
student_alcohol.applymap(times10).head(10)
```

##United States - Crime Rates - 1960 - 2014


Step 1. Import the necessary libraries


```python
import numpy as np
import pandas as pd
```

Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv).


Step 3. Assign it to a variable called crime.


```python
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
crime.head(10)
```

Step 4. What is the type of the columns?


```python
crime.info()
```

Step 5. Convert the type of the column Year to datetime64


```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

Step 6. Set the Year column as the index of the dataframe


```python
crime = crime.set_index('Year', drop=True)
crime.head()
```

Step 7. Delete the Total column


```python
del crime['Total']
crime.head()
```

Step 8. Group the year by decades and sum the values


```python
crimes = crime.resample('10YS').sum()
```

```python
population = crime['Population'].resample('10YS').max()
```

```python
crimes['Population'] = population
crimes
```

Step 9. What is the most dangerous decade to live in the US?

```python
# apparently the 90s was a pretty dangerous time in the US
crimes.idxmax(0)
```
