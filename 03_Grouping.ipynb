##Ex - GroupBy


Step 1. Import the necessary libraries


```python
import pandas as pd
```

Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).


```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
```

Step 3. Assign it to a variable called drinks.(Watch the values of Column continent NA (North America), and how Pandas interprets it!


```python
drinks = pd.read_csv(url)
drinks
```

Step 4. Which continent drinks more beer on average?


```python
drinks.groupby('continent').beer_servings.mean()
```

Step 5. For each continent print the statistics for wine consumption.


```python
drinks.groupby('continent').wine_servings.describe()
```

Step 6. Print the mean alcohol consumption per continent for every column


```python
drinks.groupby('continent').mean(numeric_only=True)
```

Step 7. Print the median alcohol consumption per continent for every column


```python
drinks.groupby('continent').median(numeric_only=True)
```

Step 8. Print the mean, min and max values for spirit consumption for each Continent.


```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

##Occupation

Step 1. Import the necessary libraries


Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).


```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

Step 3. Assign it to a variable called users.

```python
users = pd.read_table(url, sep='|', index_col='user_id')
users
```

Step 4. Discover what is the mean age per occupation


```python
users.groupby('occupation').age.mean()
```

Step 5. Discover the Male ratio per occupation and sort it from the most to the least


```python
def gender_column(x):
  if x == 'M':
    return 1
  if x == "F":
    return 0

users['gender_n'] = users.gender.apply(gender_column)
a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts()*100
a.sort_values(ascending=False)
```

Step 6. For each occupation, calculate the minimum and maximum ages


```python
users.groupby('occupation').age.agg(['min','max'])
```

Step 7. For each combination of occupation and gender, calculate the mean age


```python
users.groupby(['occupation', 'gender']).age.mean()
```

Step 8. For each occupation present the percentage of women and men


```python
gender_occup = users.groupby(['occupation', 'gender']).agg({'gender':'count'})
occup_count = users.groupby(['occupation']).agg('count')
occup_gender = gender_occup.div(occup_count, level='occupation')*100
occup_gender.loc[:,'gender']
```

##Regiment


Step 1. Import the necessary libraries


```python
import pandas as pd
```

Step 2. Create the DataFrame with the following values:


```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

Step 3. Assign it to a variable called regiment.


Don't forget to name each column


```python
regiment = pd.DataFrame(raw_data, columns = raw_data.keys())
regiment
```

Step 4. What is the mean preTestScore from the regiment Nighthawks?


```python
nighthawks_data = regiment[regiment['regiment'] == 'Nighthawks']
mean_score = nighthawks_data.groupby('regiment')['preTestScore'].mean()
print(f"The mean preTestScore for Nighthawks is: {mean_score}")
```

Step 5. Present general statistics by company


```python
regiment.groupby('company').describe()
```

Step 6. What is the mean of each company's preTestScore?


```python
regiment.groupby(['regiment','company']).preTestScore.mean()
```

Step 8. Present the mean preTestScores grouped by regiment and company without heirarchical indexing


```python
regiment.groupby(['regiment','company']).preTestScore.mean().unstack()
```

Step 9. Group the entire dataframe by regiment and company

```python
numeric_data = regiment[['regiment', 'company','preTestScore', 'postTestScore']]
mean_scores_by_group = numeric_data.groupby(['regiment', 'company']).mean()
print(mean_scores_by_group)
```

Step 10. What is the number of observations in each regiment and company


```python
regiment.groupby(['company', 'regiment']).size()
```

Step 11. Iterate over a group and print the name and the whole data from the regiment


```python
for name, group in regiment.groupby('regiment'):
  print(name)
  print(group)
```
